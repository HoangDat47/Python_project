#### Ép kiểu dữ liệu

```python
#Chuyển kiểu string sang ngày tháng
df.date_taken = pd.to_datetime(df.date_taken)
```

```python
#Chuyển các cột sang kiểu float
df[['Q6 - How Happy are you in your Current Position with the following? (Salary)','Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)']]=df[['Q6 - How Happy are you in your Current Position with the following? (Salary)','Q6 - How Happy are you in your Current Position with the following? (Work/Life Balance)']].astype(float)
```

```python
df.age = df.age.astype(np.int16)
```

#### Xử lý dữ liệu

#### Thay rỗng bằng NA

```python
df = df.replace(r'^s*$',np.nan)
```

```python
#Kiểm tra DL trong cột 'current_role'
df['curent_role'].unique()
#df['curent_role'].nunique()
```

#### Cột 'curent_role' chứa rất nhiều giá trị khác hau. Gom các Other... thành 1 loại other

```python
df['curent_role']=df['curent_role'].str.replace('Other.*','Other',regex=True)
```

```python
df['curent_role'].unique()
```

```python
df.columns
```

```python
#Kiểm tra cột salary
```

```python
df['salary'].unique()
```

#### Tách cột salary thành 2 cột khác nhau

```python
df[['sal_min','sal_max']]=df['salary'].str.split('-',n=2,expand=True)
```

```python
df.head(1)
```

```python
df['sal_min']=df['sal_min'].str.extract('(\d+)')
df['sal_max']=df['sal_max'].str.extract('(\d+)')
```

```python
df.head(1)
```

```python
df['sal_max'].unique()
```

```python
#Trường hợp sal_max = NA vi 225k+
df.loc[df['sal_max'].isna(),'sal_max']=df['sal_min']
```

```python
df['sal_max'].unique()
```

```python
df[['sal_max','sal_min']]=df[['sal_max','sal_min']].astype(float)
```

```python
df.info()
```

```python
df['salary']=(df['sal_min']+df['sal_max'])/2
#sau đó thì drop 2 cột min max
```

```python
df['language']=df['language'].str.replace('Other.*','Other',regex=True)
```

```python
df.loc[df['age']>100,'age']
df['age'].mean()
```

```python
#Lấy dữ liệu ngẫu nhiên
df.sample(5)
```

```python
df.loc[df['age']>100, 'age'] = df['age'].mean()
```

```python
#tao ban sao
df_clean = df.copy()
```

```python
#luu len dia
df_clean.to_csv("job_survey_clean.csv")
```

#### Thong ke theo ngon ngu lap trinh

```python
df_language = df_clean.groupby('language').agg({'id':'count','salary':'mean'}).reset_index()
```

```python
df_language
```

```python
df_language.plot(kind = "bar",x='language')
plt.grid()
plt.title('So luong nguoi dung va muc luong tung ngon ngu LT')
```

```python
df_language.plot(kind = "barh",x='language')
plt.grid()
plt.title('So luong nguoi dung va muc luong tung ngon ngu LT')
```

Đối với lương trung bình của các nhóm ngôn ngữ. Từ plot chúng ta thấy rằng java có mức lương trung bình cao nhất. Tuy nhiên vì số lượng người dùng java chỉ có 1 nên rất khó để đánh giá chính xác mức lương của java. Nếu loại java ra khỏi phân tích thì ra thấy R là nhóm ngôn ngữ có trung bình lương cao nhất (số).

#### Thong ke theo cong viec

```python
df_role=df_clean.groupby('curent_role').agg({'id':'count','salary':'mean'}).reset_index()
```

```python
df_role
```

```python
df_role.plot(kind='bar',x='curent_role',y='id')
plt.xticks(rotation=30) #Quay nhan truc x 30 nghieng do
plt.title('Polular jobs') #Tieu de do thi
plt.xlabel('Job title') #Tieu de truc x
plt.ylabel('Number oof takers') #Tieu de truc y
#addlabels('Popular jobs','Job title','Number oof takers')
#adddata(df_role.curent_role,df_role.id)
plt.grid() #Ke luoi do thi
```

```python
df_clean['age'].plot.hist(bins=20) #bin nhom du lieu thanh 20 nhom
plt.xticks(rotation=30)
plt.xlabel('Age')
plt.ylabel('Number of takers')
plt.title('Working age')
plt.grid()
```

Có 1 người ở độ tuổi 90 nhà tuyển dụng nên xem xét lại trường hợp này có thể sai sót hoặc là quá tuổi.

#### Mức độ phổ biến của các ngôn ngữ tương ứng trong công việc

```python
df_job_language=df_clean.groupby(['curent_role','language']).agg({'id':'count','salary':'mean','age':'mean'}).reset_index()
```

```python
plt.figure(figsize=[6,6])
sb.barplot(data=df_job_language, x='curent_role',y='id',hue='language') #,palette='viridis'
plt.xticks(rotation=30)
plt.xlabel('Job')
plt.ylabel('Count of takers')
plt.title('Popular language')
plt.grid()
```
