## 3. Clean data

```python
#Make copies of the datasets to ensure the raw dataframes are not impacted
patients_copy = patients.copy()
treatments_copy = treatments.copy()
```

### Quality Issue 1: Null data

```python
patients_copy = patients_copy.dropna(axis=0)
```

```python
patients_copy.isna().sum()
```

```python
mean = treatments_copy['hba1c_change'].mean()
treatments_copy['hba1c_change'].fillna(mean, inplace=True)
```

```python
treatments_copy.isna().sum()
```

### Tidiness Issue 1: Unnecessary columns

```python
patients_copy.drop(['address','city','state', 'zip_code', 'country', 'contact'],axis=1)
```

```python
# Tiêu chuẩn hóa các cột tên sang chữ thường
patients_copy['given_name'] = patients_copy['given_name'].str.lower()
patients_copy['surname'] = patients_copy['surname'].str.lower()
treatments_copy['given_name'] = treatments_copy['given_name'].str.lower()
treatments_copy['surname'] = treatments_copy['surname'].str.lower()
```

### Remove unnecessary variables and combine datasets if needed

```python
#Merge 2 dataset
df_merge = pd.merge(patients_copy, treatments_copy, on=["given_name", "surname"])
df_merge
```

```python
df_merge.to_csv('Merged.csv', index=False)
```
