```python
# Tạo biểu đồ thống kê giá trung bình của từng quốc gia
plt.figure(figsize=(20, 6))
ax = sns.barplot(x='COUNTRY', y='PRICEEACH', data=df, estimator=lambda x: sum(x) / len(x))
plt.title('Giá trung bình của từng quốc gia')
plt.xlabel('Quốc gia')
plt.ylabel('Giá trung bình')

# Hiển thị giá trị cụ thể trên các cột
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')
plt.show()
```

```python
# Nhóm dữ liệu theo 'PRODUCTLINE' và 'COUNTRY', và đếm số lượng
popular_productlines = df.groupby(['PRODUCTLINE', 'COUNTRY']).size().reset_index(name='COUNT')

# Sắp xếp kết quả theo số lượng từ cao đến thấp
popular_productlines = popular_productlines.sort_values(by='COUNT', ascending=False)

# Chọn 5 quốc gia có số lượng sản phẩm được ưa chuộng cao nhất
top_countries = popular_productlines['COUNTRY'].unique()[:5]

# Lọc dữ liệu cho các quốc gia này
top_productlines = popular_productlines[popular_productlines['COUNTRY'].isin(top_countries)]

# Vẽ biểu đồ cột
plt.figure(figsize=(10, 6))
for country in top_countries:
    country_data = top_productlines[top_productlines['COUNTRY'] == country]
    plt.bar(country_data['PRODUCTLINE'], country_data['COUNT'], label=country)

plt.xlabel('Product Line')
plt.ylabel('Count')
plt.title('Popular Product Lines by Country')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Lọc dữ liệu cho các nước thuộc khu vực châu Á
asia_countries = ['Japan', 'Philippines', 'Singapore']
asia_data = df[df['COUNTRY'].isin(asia_countries)]

# Nhóm dữ liệu theo 'PRODUCTLINE', 'PRICEEACH' và 'COUNTRY', và tính tổng doanh số
sales_by_product_price_country = asia_data.groupby(['PRODUCTLINE', 'PRICEEACH', 'COUNTRY']).agg({'SALES': 'sum'}).reset_index()

# Vẽ biểu đồ cột
plt.figure(figsize=(12, 8))
for country in asia_countries:
    country_data = asia_data[asia_data['COUNTRY'] == country]
    plt.bar(country_data['PRODUCTLINE'] + ' (' + country + ')', country_data['SALES'], label=country)

plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.title('Sales of Product Lines by Country in Asia')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
